```python
import pandas as pd
import numpy as np
```

## Houston

```python
df = pd.read_csv("./KHOU.csv")
df.head()
df
```

```python
#split data by years 2014 and 2015
fourteens = [] 
fifteens = []
for i in range(df.shape[0]):
    fourteens.append(df.date[i][:4] == '2014')
    fifteens.append(df.date[i][:4] == '2015')
data2014 = df[fourteens]
data2015 = df[fifteens]
data2015.head()
```

```python
#get the average temp for each month in each year (2014)

#print(data2014.date[0][5:7])
##july
vals7 = []
for i in range(len(data2014)):
    vals7.append(data2014.date[i][5:6] == '7')
jul = data2014[vals7]
julAvg = jul.actual_mean_temp.mean() 


##aug
vals8 = []
for i in range(len(data2014)):
    vals8.append(data2014.date[i][5:6] == '8')
aug = data2014[vals8]
augAvg = aug.actual_mean_temp.mean() 


##sep
vals9 = []
for i in range(len(data2014)):
    vals9.append(data2014.date[i][5:6] == '9')
sep = data2014[vals9]
sepAvg = sep.actual_mean_temp.mean() 

##october
vals1 = []
for i in range(len(data2014)):
    vals1.append(data2014.date[i][5:7] == '10')
octo = data2014[vals1]
octoAvg = octo.actual_mean_temp.mean() 

##nov
vals1 = []
for i in range(len(data2014)):
    vals1.append(data2014.date[i][5:7] == '11')
nov = data2014[vals1]
novAvg = nov.actual_mean_temp.mean() 

vals1 = []
for i in range(len(data2014)):
    vals1.append(data2014.date[i][5:7] == '12')
dec = data2014[vals1]
decAvg = dec.actual_mean_temp.mean() 

print('july =', julAvg,'aug =', augAvg, 'sep =', sepAvg, 'oct =', octoAvg,'nov =', novAvg, 'dec =', decAvg )
```

```python
#get the average temo for each month in each year (2015)
##january
vals1 = []
data2015.index = range(len(data2015.index))
for i in range(len(data2015)):
    vals1.append(data2015.date.loc[i][5:6] == '1')
jan15 = data2015[vals1]
janAvg15 = jan15.actual_mean_temp.mean() 

##feb
vals2 = []
for i in range(len(data2015)):
    vals2.append(data2015.date.loc[i][5:6] == '2')
feb15 = data2015[vals2]
febAvg15 = feb15.actual_mean_temp.mean() 

##march
vals3 = []
for i in range(len(data2015)):
    vals3.append(data2015.date.loc[i][5:6] == '3')
mar15 = data2015[vals3]
marAvg15 = mar15.actual_mean_temp.mean() 

##apr
vals4 = []
for i in range(len(data2015)):
    vals4.append(data2015.date.loc[i][5:6] == '4')
apr15 = data2015[vals4]
aprAvg15 = apr15.actual_mean_temp.mean() 

##may
vals5 = []
for i in range(len(data2015)):
    vals5.append(data2015.date.loc[i][5:6] == '5')
may15 = data2015[vals5]
mayAvg15 = may15.actual_mean_temp.mean() 

##jun
vals6 = []
for i in range(len(data2015)):
    vals6.append(data2015.date.loc[i][5:6] == '6')
jun15 = data2015[vals6]
junAvg15 = jun15.actual_mean_temp.mean() 

print('jan =', janAvg15,'feb =', febAvg15,'mar =', marAvg15,'apr =', aprAvg15,'may =', mayAvg15, 'jun =', junAvg15)
```

```python
#prec = actual_precipitation.mean()

##july
values = []
for i in range(len(data2014)):
    values.append(data2014.date[i][5:6] == '7')
julPrec = data2014[values]
julPrecAvg = julPrec.actual_precipitation.mean() 

##aug
values = []
for i in range(len(data2014)):
    values.append(data2014.date[i][5:6] == '8')
augPrec = data2014[values]
augPrecAvg = augPrec.actual_precipitation.mean() 

##sep
values = []
for i in range(len(data2014)):
    values.append(data2014.date[i][5:6] == '9')
sepPrec = data2014[values]
sepPrecAvg = sepPrec.actual_precipitation.mean() 

##octo
values = []
for i in range(len(data2014)):
    values.append(data2014.date[i][5:7] == '10')
octoPrec = data2014[values]
octoPrecAvg = octoPrec.actual_precipitation.mean() 

##nov
values = []
for i in range(len(data2014)):
    values.append(data2014.date[i][5:7] == '11')
novPrec = data2014[values]
novPrecAvg = novPrec.actual_precipitation.mean() 

##dec
values = []
for i in range(len(data2014)):
    values.append(data2014.date[i][5:7] == '12')
decPrec = data2014[values]
decPrecAvg = decPrec.actual_precipitation.mean() 

print('july prec=', julPrecAvg*100,'aug prec=', augPrecAvg*100, 'sep prec=', sepPrecAvg*100, 'oct orec=', octoPrecAvg*100, 'nov prec=', novPrecAvg*100, 'dec prec=', decPrecAvg*100)
```
